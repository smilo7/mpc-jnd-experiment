```python
import numpy as np
import matplotlib.pyplot as plt
import librosa

def calculate_spectral_centroid(file_path):
    """
    Calculate the spectral centroid for a given WAV file.

    Parameters:
    file_path (str): Path to the WAV file

    Returns:
    float: Spectral centroid value
    """
    # Load the audio file
    y, sr = librosa.load(file_path)

    # Calculate spectral centroid
    spectral_centroids = librosa.feature.spectral_centroid(y=y, sr=sr)

    # Return the mean spectral centroid
    return np.mean(spectral_centroids)

def analyze_spectral_centroids(base_filename_pattern, file_count):
    """
    Analyze spectral centroids for a series of WAV files.

    Parameters:
    base_filename_pattern (str): Base filename pattern
    file_count (int): Number of files to analyze

    Returns:
    tuple: (file numbers, spectral centroid values)
    """
    file_numbers = []
    spectral_centroids = []

    for i in range(1, file_count + 1):
        # Format the filename with zero-padding
        filename = f"{base_filename_pattern}_{i:02d}.wav"

        try:
            centroid = calculate_spectral_centroid(filename)
            file_numbers.append(i)
            spectral_centroids.append(centroid)
            print(f"Processed {filename}: Spectral Centroid = {centroid}")
        except Exception as e:
            print(f"Error processing {filename}: {e}")

    return file_numbers, spectral_centroids

# Main execution
base_pattern = "paper_trombone_morph"
file_count = 30

# Analyze spectral centroids
file_numbers, spectral_centroids = analyze_spectral_centroids(base_pattern, file_count)
```

```python
def plot_spectral_centroids(file_numbers, spectral_centroids):
    """
    Create a plot of spectral centroids.

    Parameters:
    file_numbers (list): List of file numbers
    spectral_centroids (list): List of corresponding spectral centroid values
    """
    plt.figure(figsize=(12, 6))
    plt.plot(file_numbers, spectral_centroids, marker='o')
    plt.title('Spectral Centroids Across Trombone Samples')
    plt.xlabel('Sample Number')
    plt.ylabel('Spectral Centroid')
    plt.grid(True)
    plt.tight_layout()
    plt.show()


# Plot the results
plot_spectral_centroids(file_numbers, spectral_centroids)
```
